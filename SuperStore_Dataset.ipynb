##Using GroupBy to get stats from correlations.

```python
df.columns
```

```python
discounts= df.groupby('Discount')['Discount'].sum()
discounts=pd.DataFrame(discounts)
discounts
```

### Get Sales per State

```python
df.groupby('State')['Sales'].sum().sort_values(ascending=False).plot(kind='bar',color='red')
```

California and New York are the highest sales.

### Get Profit per State

```python
df.groupby('State')['Profit'].sum().sort_values(ascending=False).plot(kind='bar',color='green')
```

**From both previous plots, it is concluded that New York is more profitable than California.**


###Get the highest customers in New York

```python
df[df['State'] == 'New York'].groupby('Customer Name')['Profit'].sum().sort_values(ascending=False).head(5)
```

Tom Ashbrook is the most profitable customer in New York.
